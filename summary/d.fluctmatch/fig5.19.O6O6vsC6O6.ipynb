```python
def get_trend_line(x, y, xlim, is_filter=True):
    # Filtered data
    if is_filter:
        coefficients = np.polyfit(x, y, 1)
        polynomial = np.poly1d(coefficients)
        y_fit = polynomial(x)
        residuals = y - y_fit
        outlier_mask = np.abs(residuals) < 2 * np.std(residuals)
        x = x[outlier_mask]
        y = y[outlier_mask]
    ## Generate trend line
    xlim_min, xlim_max = xlim
    xlim_min = xlim_min + (xlim_max-xlim_min) * 0.1
    xlim_max = xlim_max - (xlim_max-xlim_min) * 0.1
    x_range = np.linspace(xlim_min, xlim_max, 5)
    coefficients = np.polyfit(x, y, 1)  # 1 indicates a linear fit
    polynomial = np.poly1d(coefficients)
    y_fit_from_x_range = polynomial(x_range)
    # Calculate R^2
    y_fit = polynomial(x)
    ss_total = np.sum((y - np.mean(y))**2)  # Total sum of squares
    ss_residual = np.sum((y - y_fit)**2)         # Residual sum of squares
    r_squared = 1 - (ss_residual / ss_total)          # R^2 calculation
    info = f'{coefficients[0]:.2f}x{coefficients[1]:+.2f}, $\\rm{{R}}^{{2}}$={r_squared:.2f}'
    return x_range, y_fit_from_x_range, info
```

```python
## get all O6K in 'ks1' variable, and get all C6K in 'ks2'
gstrands_i = b_agent.df_tetrad_geometry.columns.values
layers_ij2metals   = {'top': ['K1'], 'mid':['K1','K2'], 'bot':['K2']}
dict_position = {}
ks1, ks2 = [], []
for layer, metals in layers_ij2metals.items():
    for metal in metals:
        for gstrand_i in gstrands_i:
            resid_i = b_agent.df_tetrad_geometry.loc[layer,gstrand_i]
            resid_j = 3 - int(metal[1]) ## Note: K1(resid_j=2, on top), K2(resid_j=1, on bot)
            ## generate dict_hotspot
            mask1 = df['Resid_i']==resid_i
            mask2 = df['Resid_j']==resid_j
            mask3 = df['Atomname_j']=='POT'
            mask_O6 = df['Atomname_i']=='O6'
            mask_C6 = df['Atomname_i']=='C6'
            df_maskO6 = df[mask1 & mask2 & mask3 & mask_O6]
            df_maskC6 = df[mask1 & mask2 & mask3 & mask_C6]
            ks1.append(df_maskO6['k'].squeeze())
            ks2.append(df_maskC6['k'].squeeze())
ks1, ks2 = np.array(ks1), np.array(ks2)
```

```python
p_agent = SinglePlotAgent(figsize_1=[2.5, 2.5])
p_agent.axs.scatter(ks1, ks2, c='black', alpha=0.5, s=15)
p_agent.set_xlabel(r'$k^{\rm{bm}}_{{\rm{O6K1}}}$ or $k^{\rm{bm}}_{{\rm{O6K2}}}$')
p_agent.set_ylabel(r'$k^{\rm{bm}}_{{\rm{C6K1}}}$ or $k^{\rm{bm}}_{{\rm{C6K2}}}$')
x_line, y_line, info = get_trend_line(ks1, ks2, p_agent.axs.get_xlim())
p_agent.fig.text(x=0.27, y=0.8, s=info)
p_agent.axs.plot(x_line, y_line, c='black', alpha=0.2)
p_agent.savefig('g4dna_O6K_C6K_corr.svg')
```
