### make a centered reference structure

````python
def rmpbc(filename,indexfile='index.ndx',center=0,output=0):
    file = ".".join(filename.split('.')[:-1])
    extend = filename.split('.')[-1]
    cmd1 = f'echo "{center} {output}" | gmx trjconv -s {file}.tpr -f {file}.{extend}   -o {file}_c.{extend} -pbc res   -center -boxcenter zero -n {indexfile} -ndec 5 -ur compact'
    print("##========================command=line========================##")
    print(cmd1,sep="\n")
    print("##======================end=command=line======================##")
    !{cmd1} 
````

````python
rmpbc("npt2.pdb", indexfile, center=2)
````

````python
!echo "2 2" | gmx trjconv -s rms.tpr -f heat_c.xtc -o heat_rms.xtc -n {indexfile} -fit rot+trans -ndec 5
!echo "2 2" | gmx trjconv -s rms.tpr -f nvt1_c.xtc -o nvt1_rms.xtc -n {indexfile} -fit rot+trans -ndec 5
!echo "2 2" | gmx trjconv -s rms.tpr -f npt1_c.xtc -o npt1_rms.xtc -n {indexfile} -fit rot+trans -ndec 5
!echo "2 2" | gmx trjconv -s rms.tpr -f npt2_c.xtc -o npt2_rms.xtc -n {indexfile} -fit rot+trans -ndec 5
````
